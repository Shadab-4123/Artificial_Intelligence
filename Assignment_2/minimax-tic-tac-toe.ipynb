```python
def Terminal(state):
    if state[0] == state[1] == state[2] == 'O':
        return (True,1)
    elif state[3] == state[4] == state[5] == 'O':
        return (True,1)
    elif state[6] == state[7] == state[8] == 'O':
        return (True,1)
    elif state[0] == state[3] == state[6] == 'O':
        return (True,1)
    elif state[1] == state[4] == state[7] == 'O':
        return (True,1)
    elif state[2] == state[5] == state[8] == 'O':
        return (True,1)
    elif state[0] == state[4] == state[8] == 'O':
        return (True,1)
    elif state[2] == state[4] == state[6] == 'O':
        return (True,1)
    

        
    elif state[0] == state[1] == state[2] == 'X':
        return (True,-1)
    elif state[3] == state[4] == state[5] == 'X':
        return (True,-1)
    elif state[6] == state[7] == state[8] == 'X':
        return (True,-1)
    elif state[0] == state[3] == state[6] == 'X':
        return (True,-1)
    elif state[1] == state[4] == state[7] == 'X':
        return (True,-1)    
    elif state[2] == state[5] == state[8] == 'X':
        return (True,-1)
    elif state[0] == state[4] == state[8] == 'X':
        return (True,-1)
    elif state[2] == state[4] == state[6] == 'X':
        return (True,-1)
    
 
    for row in state:
        if row == 'N':
            return (False,0)
        
    return (True,0)
```

```python
def Is_Terminal(state):
    return Terminal(state)[0]
```

```python
def win_score(state):
    return Terminal(state)[1]
```

```python
def change_state(state, position, player):
    state1 = state[:]
    new_state = ''
    for i in range(len(state)):
        if state[i] != 'O' and  state[i] != 'X' and i == position:
            new_state += player
        else:
            new_state += state1[i]
    
    return new_state                
        
```

```python
def play_game(state, player):
    if Is_Terminal(state):        
        win = {
            'state': state,
            'player': player,
            'score': win_score(state)
        }
        return win 


    record = {
        'state': state,
        'player': player,
        'score': None,
        'children': []
    }      

    opponent = 'O' if player == 'X' else 'X'

    for i, j in enumerate(state):
        if j == 'N':
            state1 = change_state(state, i, player)
            child_result = play_game(state1, opponent)
#             print(child_result)
            record['children'].append(child_result)
    
    max_score = float('-inf')
    min_score = float('inf')
    if record['player'] == 'O':
        for i in record['children']:
#             print(i)
#             break
            if max_score <= i['score']:
                max_score = i['score']
        record['score'] = max_score
    elif record['player'] == 'X':        
        for i in record['children']:
            if min_score >= i['score']:
                min_score = i['score']
        record['score'] = min_score       
       

    return record


# Example usage
k = play_game('NNNNNNNNN', 'O')
```

```python

def computer_turn(state,player):
    best_move_list = []
    best_move = state

    for i, j in enumerate(state):
        if j == 'N':
            state1 = change_state(state, i, player)            
            best_move_list.append(state1)
    
    max_score = float('inf')
    for move in (best_move_list):
        if move in parent:
            if max_score > parent[move]:
                max_score = parent[move]
                best_move = move
    return best_move            
            
```
